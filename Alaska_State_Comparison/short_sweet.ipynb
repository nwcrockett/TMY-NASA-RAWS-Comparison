# Breakdown of NASA POWER data sources

## 1 July, 1983 through 31 December, 2007
1. Source: NASA/GEWEX SRB
2. [GEWEX documention](https://gewex-srb.larc.nasa.gov/)
3. Data Production Source: [International Satellite Cloud Climatology Project (ISCCP)](https://isccp.giss.nasa.gov/) and [Global Modeling and Assimilation Office (GMAO)](https://gmao.gsfc.nasa.gov/)
Data is then reanalysised with radiative transfer algorthims.
shortwave - Pinker and Laszlo (Modeling Surface Solar Irradiance for Satellite Applications on a Global Scale, J. Appl. Met., 31, 194-211, 1992)
longwave - Fu et al., (JAS, Vol. 54, 2799-2812, 1997). The Langley Parameterized SW algorithm (LPSA; Gupta et al. - NASA/TP-2001-211272, Dec. 2001)
4. Attempted to gain accuracy of 10W m^2 or 5%. Whichever is smaller
5. Known issues: snow/ice covered surfaces and large areas of burning biomass
6. Validated using ground measurements. From the following sources: Baseline Surface Radiation Network (BSRN), the Swiss Federal Institute of Technology's Global Energy Balance Archive (GEBA) and NOAA's Climate Montitoring and Diagnostics Laboratory (CMDL)
7. Average error rate for both shortwave and longwave is 10W m^2


## FLASHFlux version 2 (2008-2013) and v3 (2013-near real time)
1. Created to speed up data processing to near realtime resolution compared to the Clouds and the Earth's Radiant Energy Systems  (CERES) and Moderate-Resolution Imaging Spectroradiometer top-of-atmosphere (MODIAS TOA) observations
2. [Data process](https://ceres.larc.nasa.gov/products.php?product=FLASHFlux-Level2)
![Flashflux process](Flashflux_process.png "Flashflux process") 
3. This data does not use the CERES instrument calbration so contains higher uncertainties. CERES takes 6 months to update new data
4. Has resolution of 0.5 degrees Longitude and Latitude that is derived from 1 degree

# TMY3 data sources and methodology
1. Data is seperated into 3 classes. Class 1 has the lowest uncertty of lower than 11%, class 2 has a higher uncertainty of greater than 11%, class three has missing data
2. Only measured data site in Alaska in Barrow using ARM data
3. The models used were the NREL Meteorological-Statistical (METSTAT) and a SUNY model that used Geostationary Operational Environmental Satellite (GOES) imagery to estimate solar radiation.
4. GOES has a 1-km resolution that are downsampled by SUNY to 10-km
5. SUNY is not used in Alaska due to Latitude above 60 degrees
